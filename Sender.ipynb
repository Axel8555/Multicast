```python
import socket
import time
import threading
import sys
import math

MCAST_GRP = "224.1.1.1"
MCAST_PORT = 5007
# regarding socket.IP_MULTICAST_TTL
# ---------------------------------
# for all packets sent, after two hops on the network the packet will not
# be re-sent/broadcast (see https://www.tldp.org/HOWTO/Multicast-HOWTO-6.html)
MULTICAST_TTL = 2

sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM, socket.IPPROTO_UDP)
sock.setsockopt(socket.IPPROTO_IP, socket.IP_MULTICAST_TTL, MULTICAST_TTL)
print("Sockets creados")
# For Python 3, change next line to 'sock.sendto(b"robot", ...' to avoid the
# "bytes-like object is required" msg (https://stackoverflow.com/a/42612820)
filename = "dspic.pdf"
filename = input()

while 1:
    print("Preparandose para enviar archivos")
    dir = "IN/"
    
    in_file = open(dir+filename, "rb") # opening for [r]eading as [b]inary
    data = in_file.read() 
    in_file.close()
    filedata = "|"+filename+"|"+str(len(data))
    while len(filedata)<256:
        filedata = filedata + '|'
    
    sock.sendto(filedata.encode(), (MCAST_GRP, MCAST_PORT))
    print("Datos del archivo enviados")
    
    if int(len(data)/256)<(len(data)/256):
        packages = math.floor(len(data)/256)+1
    else:
        packages = math.floor(len(data)/256)
        
    for i in range(packages):
        buffer = data[i*256 : min(len(data),(i+1)*256)]
        try:
            sock.sendto(buffer, (MCAST_GRP, MCAST_PORT))
        except:
            print("Fallo al enviar un paquete de datos: %s/%s"%(i+1, packages))
            break;
        print("%s/%s"%(i+1, packages))
    else:
        print("Archivo enviado correctamente")
    
    print("Enviando archivos nuevamente en:")
    for i in range(1,10):
        print(10-i)
        time.sleep(1)
    
    #filename = input()
```
